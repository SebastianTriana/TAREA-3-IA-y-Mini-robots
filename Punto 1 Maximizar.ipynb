```python
import math
import random
import numpy as np

mejor_crom = None
mejor_x = None
mejor_fit = -1e9
mejor_gen = 0

##decodifica un cromosoma
def decodifica(crom, l):
    valor_entero = int("".join(str(int(bit)) for bit in crom), 2)
    x = valor_entero / (2**l - 1)
    return x

def ecuacion(x):
    return 10*(x * math.sin(10 * math.pi * x) + 1) ##funcion de aptitud

def evalua(crom, l):
    x = decodifica(crom, l)
    y = ecuacion(x) ##funcion de aptitud es igual a la misma funcion a maximizar
    return y

def generar_cromosoma(l):
    return [random.randint(0, 1) for _ in range(l)]## cromosomas aleatorios
```

```python
def next_generation (poblacion,N):
  L = np.size(poblacion[0])
  mutacion = 1
  prob_mutacion = mutacion/(N*L)
  aptitud = [evalua(x,L) for x in poblacion]
  aptitud = [x/sum(aptitud) for x in aptitud]
  new_generation = []
  ##ruleta aleatoria _  cruce
  while len(new_generation) <= N:
    for i in range (L):
      if random.random() <= aptitud[i] and len(new_generation) <= N:
        new_generation.append(poblacion[i])

  ##Cruce
  sons_generation = []
  for i in range(N//2):
    swipe = int(random.random()*L)
    son_1 = np.concatenate((new_generation[2*i][:swipe], new_generation[2*i + 1][swipe:]))
    son_2 = np.concatenate((new_generation[2*i + 1][:swipe], new_generation[2*i][swipe:]))
    sons_generation.append(son_1)
    sons_generation.append(son_2)

  ##Mutacion
  for i in range (N):
    for j in range (L):
      if random.random() < prob_mutacion:
        sons_generation[i][j] = 1 - sons_generation[i][j]

  return sons_generation
```

```python
def search_best_crom(poblacion, L,Gen):
  global mejor_crom
  global mejor_x
  global mejor_fit
  global mejor_gen

  for crom in poblacion:
    fit= evalua(crom, L)
    x = decodifica(crom,L)
    if fit >= mejor_fit:
      mejor_fit = fit
      mejor_crom = crom
      mejor_x = x
      mejor_gen = Gen+1
  print(f"\nGeneracion actual:{Gen+1}")
  print(f"=== Mejor cromosoma total: de la generacion {mejor_gen} ===")
  print("Cromosoma:", mejor_crom)
  print("x* =", mejor_x)
  print("f(x*) =", mejor_fit/10)
```

```python
def Ag(L = 10, N = 100, Gen = 1000):
  poblacion = [generar_cromosoma(L) for _ in range(N)] ##dise√±amos la poblacion original
  for generacion in range (Gen):
    search_best_crom(poblacion,L,generacion)
    poblacion = next_generation(poblacion,N)
Ag()
```

Como podemos ver, la aplicacion del algoritmo genetico permite encontrar el maximo posible de la funcion por medio de l a optimzacion de los coromosomas. Alcanzando el maximo teorico posible con 1000 generaciones y 100 individuos.
